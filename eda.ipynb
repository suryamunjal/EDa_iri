#univariate analysis
#1 d plot 

```python
for i in df.columns[:-1]:
    print(i)
    sns.FacetGrid(df, hue="species", size=5) \
    .map(sns.distplot,i) \
    .add_legend();
    plt.show();
```

petal length and petal width seems to be an important feature

#bivariate analysis
#2d plot --- use any 2 features

```python
#scatter plot using seaborn
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="species", size=4) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend();
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="species", size=4) \
   .map(plt.scatter, "petal_length", "petal_width") \
   .add_legend();
plt.show()
```

#pair plots

```python
#we cannot visualize 3d or 4d or higher dimension/features
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df, hue="species", size=3);
plt.show()
```

petal width and petal length are good feature. 

setosa can be easily separated.

#CDF : CUMMULATIVE DENSITY FUNCTION

```python
counts, bin_edges = np.histogram(df[df['species']=="versicolor"]['petal_width'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
#print(pdf);
#print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(df[df['species']=="versicolor"]['petal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
```
